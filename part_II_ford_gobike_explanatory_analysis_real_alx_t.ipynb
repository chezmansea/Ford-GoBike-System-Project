```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas datafram
ford_bike = pd.read_csv('ford_bike_df')
ford_bike.head(10)
```

```python
#Lets have a datatable for the top 10 start stations
top10_start_station = ford_bike['start_station_name'].value_counts().index[:10]
top10_start_station_df = ford_bike.loc[ford_bike['start_station_name'].isin(top10_start_station)]
```

```python
top10_start_station_df
```

### DISTRIBUTION OF AGE BY GENDER AND USER TYPE FOR THE TOP 10 START STATIONS

>Age distribution across gender show a similar trend with most riders falling within 30 to 40 years.
> Also noticeable is that subscribers recorded more trips than customers

```python
bins = np.arange(0, top10_start_station_df['age'].max()+3, 10)
g = sb.FacetGrid(data=top10_start_station_df, row = 'user_type', col= 'member_gender', margin_titles=True)
g.map(sb.histplot, 'age', bins=bins)
plt.title('AGE DISTRIBUTION BY GENDER AND USER TYPE');
```

## TRIP DURATION BY USER TYPES FOR TOP 10 START STATIONS

> Surprisingly, customers are having a longer trip duration compared to subscribers. We can also see a regular trend of a little over 500 seconds for trips recorded for subscribers in all the top 10 start stations as against the irregular pattern observed in customers trip records in the top 10 start stations.
> The irregular trend observed for customers is possibly because most of the customers are one off users going to different destinations while most of the subscribers would likely have a fixed destination they go everyday like their place of work.

```python

sb.barplot(data=top10_start_station_df, y='start_station_name', x='duration_sec', ci=None, hue='user_type', orient='h')
plt.title('Start Station By User Type Trip duration (sec)', fontsize=15)
plt.legend(['Customer', 'Subscriber'], title = 'User Type')
plt.xlabel('Duration(seconds)', fontsize=15)
plt.ylabel('Start Station Names', fontsize=15);
```

## AGE RANGE AND  TRIP DURATION BY USER TYPES

A negative correlation is observed between trip duration and age range.
This trend is similar for both customers and subscribers

```python
base_color = {'subscriber':'#99ee99', 'customer':'#d799ee'}
g = sb.FacetGrid(data = ford_bike, height = 4,col = 'user_type', size= 6, palette=base_color)
g.map(plt.scatter, 'age', 'duration_sec')
g.set_ylabels('Duration(Seconds)', size=10)
g.set_xlabels('Age Range(years)', size=10)
plt.title('AGE RANGE AND TRIP DURATION BY USER TYPES', fontsize=19)
plt.legend(['Subscriber', 'Customer'], title='User Types')
plt.subplots_adjust(top=0.6);
```
